```python
# @Course: CS_121 #19990SP26C
# @author: Jacob Yeung #900494756
# @Date: 20260411


import pandas as pd

FILENAME = 'paralympics_2024_medal_table.csv'
df = pd.read_csv(FILENAME)

Gold = "Gold"
Sliver = "Silver"
Bronze = "Bronze"
Country = "Country"
# the country has more gold than silver
dfMostGoldThanSliver = df[df[Gold] > df[Sliver]]
# the other has more gold than bronze
dfMostGoldThanBronze = df[df[Gold] > df[Bronze]]


dfMostGoldThanSliver.to_csv("dfMostGoldThanSliver.csv", index=False)
dfMostGoldThanBronze.to_csv("dfMostGoldThanBronze.csv", index=False)

listOfTheCountriesAppearInBoth = list(set(dfMostGoldThanSliver[Country]) & set(dfMostGoldThanBronze[Country]))

print("The countries that appears in Both:")
for country in listOfTheCountriesAppearInBoth:
    print(country)

print("Completed")
```
