## 01 Function signature: Defines rolling_average(input_series, period) and returns a list.

```python
from decimal import Decimal

# 01 Function signature: Defines rolling_average(input_series, period) and returns a list.
# def rolling_average(input_series: list[Decimal|int|float], period: int|Decimal): -> list[Decimal|int]:
def rolling_average(input_series, period): #-> list[Decimal|int]:
    if period <= 0:
        raise ValueError("Period size must be greater than 0.")
    if isinstance(period, Decimal) and not period.is_integer(): # as_integer_ratio
        raise ValueError("Period must be an integer.")
    # period = int(period)  # Convert to integer if it's a Decimal
    if (period := int(period)) > (input_size := len(input_series)):
        raise ValueError("Period size must be less than or equal to the length of the input_series.")
    if not all(isinstance(x, (int, float, Decimal)) for x in input_series):
        raise ValueError("All elements in input_series must be numbers.")

    # Required technique: Uses list comprehension OR map + lambda (map over a range of indices).
    result = list(map(lambda x: Decimal(str(x)), input_series))
    # 02 Correct rolling average logic: Uses window slices correctly; output length is len(input_series) - period + 1.
    result = list(map(lambda i: sum(result[i:i+period]) / period, range(input_size-period+1)))
    return [int(x) for x in result] if all(x % 1 == 0 for x in result) else result
```

### Testing 01: Basic functionality with integers
