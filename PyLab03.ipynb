## Lab: 2D Moving Average (Box Blur) with NumPy ndarrays

In the last lab you computed a moving/rolling average on a 1D list. In this lab, you’ll extend that idea to 2D space using NumPy ndarrays. This is the same core idea behind a blur effects, they are basically a weighted average applied to a pixel based on the value of the pixels around it. In this lab you will be moving an array or a "box" from the start into the end of the list. One way of seeing this is a box moving around the 2d array from the beginning of the list till the end.

This idea will become more relevant when we get to machine learning and you start looking into Convolution for Image data.

#### Write a pure function that takes only a 2D NumPy array and returns a new 2D array that contains the 3×3 moving average (box blur) for every valid center cell.
** Include your function box_blur_3x3(input_array) (one parameter only).

```python
import random

# NumPy installed and imported as import numpy as np
import numpy as np

# Use NumPy features: .shape, slicing, and .mean().
def box_blur_3x3(inputArray):
    # Read input dimensions: n, m = inputArray.shape
    n, m = inputArray.shape
    # Create an output array of zeros with shape (n-2, m-2)
    outputArray = np.zeros((n - 2, m - 2))

    for i in range(1, n - 1):
        for j in range(1, m - 1):
            # Comfort with 2D indexing and slicing: A[r0:r1, c0:c1]
            # For each valid center (exclude borders), slice a 3×3 window
            window = inputArray[i-1:i+2, j-1:j+2]
            # Compute the window average using NumPy mean()
            # Store the result in the output array
            outputArray[i-1, j-1] = window.mean()

# Create and return a NumPy output array with shape (n − 2, m − 2).
    return outputArray
```

```python
import numpy as np

input_array = np.array([[3, 5, 6, 7,1],
                        [2,3,4,9,4],
                        [4,1,5,1,2],
                        [2,8,4,7,3],
                        [6,8,9,1,2]])

print("## Test 01: One small, readable array (ex: 5×5 with known values)")
np.set_printoptions(linewidth = 2000, precision=1, suppress=True)
print("Input Array:\n", input_array)
print("Blurred Array:\n", blurred := box_blur_3x3(input_array))
```
