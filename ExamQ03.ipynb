## Question 03:
Provided here is a database with birthdays through out many years. What I would like you to do is find out what is the average of births on your birthday and the average of a made up birthday based on your ID number. The month is the first digit of your student ID number, and the day is the last digit of your student ID. Tell me if you your birthday or the one based on your ID has a larger average.

```python
import pandas as pd
from fractions import Fraction

# Provided here is a database with birthdays through out many years
FILENAME = 'birthdays.csv'
df = pd.read_csv(FILENAME)

# What I would like you to do is find out what is the average of births on your birthday and the average of a made up birthday based on your ID number.
myBirthday = "11-13" # Nov 13
# The month is the first digit of your student ID number, and the day is the last digit of your student ID.
myStudentID = "900494756"
myIDBirthday = f"{myStudentID[0]}-{myStudentID[len(myStudentID)-1]}" # StudentID: 900494756
# print(myIDBirthday) # 9-6

myBirthdayArray = myBirthday.split("-")
myBirthdayArray = [int(x) for x in myBirthdayArray]
myIDBirthdayArray = myIDBirthday.split("-")
myIDBirthdayArray = [int(x) for x in myIDBirthdayArray]

# print(myBirthdayArray)
# print(myIDBirthdayArray)

data = df[['month', 'date_of_month', 'births']]
myBirthdayCount = 0
myIDBirthdayCount = 0
myBirthdaySum = 0
myIDBirthdaySum = 0
for index, row in data.iterrows():
    if row['month'] == myBirthdayArray[0] and row['date_of_month'] == myBirthdayArray[1]:
        myBirthdaySum += row['births']
        myBirthdayCount += 1
    if row['month'] == myIDBirthdayArray[0] and row['date_of_month'] == myIDBirthdayArray[1]:
        myIDBirthdaySum += row['births']
        myIDBirthdayCount += 1

# Tell me if you your birthday or the one based on your ID has a larger average.
myBirthdayAverage = Fraction(myBirthdaySum, myBirthdayCount)
myIDBirthdayAverage = Fraction(myIDBirthdaySum, myIDBirthdayCount)
print(f"My birthday average: {myBirthdayAverage} ≈ {float(myBirthdayAverage)} ")
print(f"My ID birthday average: {myIDBirthdayAverage} ≈ {float(myIDBirthdayAverage)} ")

if myBirthdayAverage == myIDBirthdayAverage:
    print("My birthday and my ID birthday have the same average")
elif myBirthdayAverage > myIDBirthdayAverage:
    print(f"My birthday has a larger average {myBirthdayAverage} than my ID birthday's average {myIDBirthdayAverage}.")
else:
    print(f"My ID birthday has a larger average {myIDBirthdayAverage} than my birthday's average {myBirthdayAverage}.")

```
